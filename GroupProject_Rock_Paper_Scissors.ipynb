Group Project: Rock, Paper, Scissors Simulation

```python
def winner(): ## Calling "winner" - see call in the last line
  player1 = input("Player 1, enter your choice: ").strip().lower() ## Strip + Lower will strip any case from the input and revert it to lowercase
  player2 = input("Player 2, enter your choice: ").strip().lower()

  if player1 == player2:
    print("Tie") ## This is what outputs when there is a tie. Everything below are other possible outcomes.
  elif player1 == "rock" and player2 == "scissors":
    print("Player 1 wins!")
  elif player1 == "rock" and player2 =="paper":
    print("Player 2 wins!")
  elif player1 == "paper" and player2 == "rock":
    print("Player 1 wins!")
  elif player1 == "paper" and player2 == "scissors":
    print("Player 1 wins!")
  elif player1 == "scissors" and player2 == "rock":
    print("Player 2 wins!")
  elif player1 == "scissors" and player2 == "paper":
    print("Player 1 wins!")
  else:
  ## What happens when the player does not enter a valid option
    print("error :( ")

winner()
```
